# Sieve of Eratosthenes

```python
def sieveOfEratosthenes(k):
    prime = [True for i in range(k + 1)]
    primeNumbers = []
    
    for i in range(2, k + 1):
        if prime[i]:
            primeNumbers.append(i)
            for j in range(i * i, k + 1, i):
                prime[j] = False
    
    return primeNumbers

print("Prime numbers from 1 to 100 are: ", sieveOfEratosthenes(100))
print("Total prime numbers from 1 to 100 are: ", len(sieveOfEratosthenes(100)))
```
